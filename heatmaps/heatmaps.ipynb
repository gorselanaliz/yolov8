```python
import cv2
from ultralytics import YOLO, solutions

model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture("test.mp4")
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Video writer
save_path = "results/heatmap_output.avi"
writer = cv2.VideoWriter(save_path,
                         cv2.VideoWriter_fourcc(*'XVID'),
                         int(fps),
                         (w,h))

# Init heatmap
heatmap_obj = solutions.Heatmap(
    colormap=cv2.COLORMAP_PARULA,
    view_img=True,
    shape="circle",
    classes_names=model.names,
)


total_fps = 0
average_fps = 0
num_of_frame = 0

video_frames = []

total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print("[INFO].. Total Frames:", total_frames)


while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break
    tracks = model.track(im0, persist=True, show=False)

    im0 = heatmap_obj.generate_heatmap(im0, tracks)

    num_of_frame += 1
    video_frames.append(im0)
    print("(%2d / %2d) Frames Processed" % (num_of_frame, total_frames))


cap.release()

for frame in video_frames:
    writer.write(frame)

writer.release()
print("[INFO].. Video is saved in "+save_path)
```

```python
### Specific Classes

import cv2
from ultralytics import YOLO, solutions

model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture("test.mp4")
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

classes_for_heatmap = [0,2]

# Video writer
save_path = "results/heatmap_output.avi"
writer = cv2.VideoWriter(save_path,
                         cv2.VideoWriter_fourcc(*'XVID'),
                         int(fps),
                         (w,h))

# Init heatmap
heatmap_obj = solutions.Heatmap(
    colormap=cv2.COLORMAP_PARULA,
    view_img=True,
    shape="circle",
    classes_names=model.names,
)


total_fps = 0
average_fps = 0
num_of_frame = 0

video_frames = []

total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print("[INFO].. Total Frames:", total_frames)


while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break
    tracks = model.track(im0, persist=True, show=False)

    im0 = heatmap_obj.generate_heatmap(im0, tracks, classes=classes_for_heatmap)

    num_of_frame += 1
    video_frames.append(im0)
    print("(%2d / %2d) Frames Processed" % (num_of_frame, total_frames))


cap.release()

for frame in video_frames:
    writer.write(frame)

writer.release()
print("[INFO].. Video is saved in "+save_path)
```
